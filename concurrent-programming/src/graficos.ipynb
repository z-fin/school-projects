```python
import matplotlib.pyplot as plt
import subprocess
```

```python
# Compilar el C++
subprocess.run(['g++', '-o', '../build/experimentos', '../src/experimentos.cpp', '../src/HashMapConcurrente.cpp', '../src/CargarArchivos.cpp'], check=True)

# Ejecutar el C++
subprocess.run(['../build/experimentos'], check=True)
```

```python
# Listas para almacenar los datos
hilos = []
tiempos = []

# Leer los datos del archivo
with open('tiempoCargar/todosLosTiempos.txt', 'r') as file:
    for line in file:
        parts = line.strip().split(", ")
        if len(parts) == 2:
            hilos.append(int(parts[0].split(":")[1].strip()))  # Extraer el número de hilos
            # Eliminar ' ms' del tiempo antes de convertirlo a int
            tiempo_str = parts[1].split(":")[1].strip()  # Tiempo en formato '120 ms'
            tiempo = int(tiempo_str.replace(' ms', '').strip())  # Eliminar ' ms'
            tiempos.append(tiempo)

# Crear el gráfico
plt.figure(figsize=(8, 6))
plt.plot(hilos, tiempos, marker='o', linestyle=' ', color='b')

# Etiquetas y título
plt.xlabel('Cantidad de hilos')
plt.ylabel('Tiempo (ms)')
plt.title('Tiempo de carga en función de la cantidad de hilos')
plt.grid(True)

# Mostrar el gráfico
plt.tight_layout()
plt.savefig("tiempoCargar/todosLosTiempos.png")  # Guardar el gráfico como imagen
plt.show()

```

```python
# Leer los tiempos desde el archivo
tiempos = []
hilos = []

with open("tiempoCargar/masHilosQueArchivos.txt", "r") as file:
    for line in file:
        parts = line.split(":")
        if len(parts) == 2:
            hilo_id = int(parts[0].strip().split()[1])  # Hilo X
            tiempo = float(parts[1].strip().split()[0])  # X segundos
            hilos.append(hilo_id)
            tiempos.append(tiempo)

# Graficar los tiempos de ejecución
plt.figure(figsize=(8, 6))
plt.bar(hilos, tiempos, color='blue')
plt.title('Tiempos de Ejecución más hilos que archivos a cargar')
plt.xlabel('Nro de Hilo')
plt.ylabel('Tiempo (segundos)')
plt.xticks(hilos)  # Asegurarse de que cada hilo tenga su etiqueta
plt.grid(axis='y')

# Mostrar el gráfico
plt.tight_layout()
plt.savefig("tiempoCargar/masHilosQueArchivos.png")  # Guardar el gráfico como imagen
plt.show()
```
